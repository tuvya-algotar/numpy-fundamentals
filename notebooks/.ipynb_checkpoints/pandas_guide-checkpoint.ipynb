```python
import pandas as pd
import numpy as np
```

```python
print("Pandas version: ",pd.__version__)
print("\nPandas is ready to use")
```

```python
# Creating a series

grades = pd.Series([85,45,93,23,80])
print("Series from list: ")
print(grades)
print("\nType : ",type(grades))
```

```python
# Method 2 - Using custom index
grades_named = pd.Series(
    [85,90,78,82,99],
    index=['Math','Physics','Chemitry','Biology','CS']
)
print("Series with custom index: ")
print(grades_named)

# Accessing a name
print("\nMath grade: ",grades_named["Math"])
print("\nCS grade: ",grades_named["CS"])
```

```python
#Method 3 - FRom a dict

student_scores = {
    'Tuvya' : 85,
    'Rahul' : 34,
    'Priya' : 56,
    'Raj' : 82
}

scores_series = pd.Series(student_scores)
print("Series from dictionary:")
display(scores_series)

print("\nRahul score: ",scores_series['Rahul'])
```

```python
# Series based operation

print("Original grades")
print(grades_named)

# Mathematical operations

print("\n+5 bonus points for everyone:")
print((grades_named+5).clip(upper=100))

print("\nDouble all grades:")
print((grades_named*2).clip(upper=200))

#Statistical operations
print("="*50)
print("STATISTICS")
print("="*50)
print(f"Mean (average): {grades_named.mean():.2f}")
print(f"Median: {grades_named.median():.2f}")
print(f"Standard deviation: {grades_named.std():.2f}")
print(f"Minimum: {grades_named.min()}")
print(f"Maximum: {grades_named.max()}")
print(f"Sum: {grades_named.sum()}")
```

```python
#Creating a DataFrame:


# From dict of lists
students_data = {
    'Name': ['Tuvya', 'Raj', 'Rahul', 'Priya', 'Arjun'],
    'Age': [18, 18, 19, 18, 19],
    'Math': [85, 92, 78, 90, 88],
    'Physics': [80, 88, 82, 85, 90],
    'Chemistry': [88, 90, 75, 92, 85]
}

df = pd.DataFrame(students_data)
print("Dataframe created")
display(df)

print("\nType: ",type(df))
print("Shape (rows,coloums):",df.shape)
```

```python
# from list of dict
student_lists = [
    {
        'Name':'Tuvya',
        'Age':18,
        'CGPA' : 8.11
    },
    {
        'Name' : 'Raj',
        'Age': 18,
        'CGPA':8
    },
    {
        'Name':'Rahul',
        'Age':17,
        'CGPA':5
        
    }
]

df2 = pd.DataFrame(student_lists)
print("Dataframe from list of dict:")
display(df2)
```

```python
# From numpy array

data = np.array([
    [85, 80, 88],
    [92, 88, 90],
    [78, 82, 75]
])

df3 = pd.DataFrame(
    data,
    columns=['Math','Physics','English'],
    index=['Tuvya','Raj','Rohan']
)

print("Dataframe from NUmpy arrays:")
display(df3)
```

```python
print("Full DataFrame:")
display(df)
print("\n" + "=" * 60)

# head() - first N rows (default 5)
print("\nFirst 3 rows:")
display(df.head(3))

# tail() - last N rows
print("\nLast 2 rows:")
display(df.tail(2))

# info() - overview of DataFrame
print("\nDataFrame info:")
display(df.info())

# describe() - statistical summary
print("\nStatistical summary:")
display(df.describe())
```

```python
# Accessing Coloums

# Method1 - Using bracket notation
print("All names")
print(df['Name'])
print("\nType:",type(df['Name']))

#MEthod2 - using dot notation(only if column nams has no space)
print("\nAll ages:")
print(df.Age)

print("\nName and Math scores:")
print(df[['Name','Math']])
```

```python
# Accessing Rows
print("First row (iloc[0]): ")
print(df.iloc[0])
print("\n",df2.iloc[0])
print("\n",df3.iloc[1])
print(f"\n{df['Name']}\n\n{df.loc[:,'Math']}")
```
