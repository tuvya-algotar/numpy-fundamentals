## 📅 Day 1: Series & DataFrame Basics

### 🎯 Objective
Understand core Pandas structures: Series and DataFrame, and basic operations.

```python
import pandas as pd
import numpy as np
print("Pandas version: ",pd.__version__)
```

```python
# Create Series (basic)
data = [10,20,30,40,50]
s = pd.Series(data)
print(s)
```

```python
# Series with custom index
marks = pd.Series([85,90,75],index=["Math","Physics","Chemistry"])
print(marks)
print("Math marks:",marks["Math"])
```

```python
# Series operations
print("Add 5 marks:\n",marks+5)
print("Mean: ",marks.mean())
print("Max: ",marks.max())
print("Min: ",marks.min())
```

```python
# Create DataFrame
data = {
    "Name" : ["Tuvya","Rahul","Priya","anag","Raj"],
    "Age" : [18,19,18,18,20],
    "Marks" : [95,90,92,80,75],
    "Department":["AI","DS","CS","AI","CS"],
    "Interests":["Gaming","Football","Cricket","Studying","Gaming"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# DataFrame basics
print("Shape:",df.shape)
print("\n")
print("Coloums:",df.columns)
print("\n")
print("Data types:\n",df.dtypes)
print("\n")
print("First five data:\n",df.head())
print("\n")
print("Last five data: \n",df.tail())
```

```python
print(df["Department"])
print("\n")
print(df[["Name","Department"]])
```

```python
print("Data of first student:\n",df.loc[0])
print("\n")
print(df.loc[1:2])
```

```python
df2 = df.set_index("Department")
print(df2)
print("\n")
print(df2.loc["CS"])
print(df2.loc["AI"])
```

```python
df["Grade"] = ["A++","A","A+","B","C"]
df["Percentage"] = (df["Marks"] / 100) * 100
print(df)
```
