## CELL 1: What is Pandas?

**Pandas** = Python Data Analysis Library

### Why Pandas?
- Built on top of NumPy (uses NumPy arrays internally)
- Designed for working with **tabular data** (like Excel spreadsheets, SQL tables, CSV files)
- Makes data cleaning, analysis, and manipulation **EASY**

### Real-world use:
- Data Science: 90% of time spent cleaning/preparing data → Pandas does this
- Machine Learning: Need to prepare data before feeding to ML models → Pandas
- Business Analytics: Analyzing sales, customer data, trends → Pandas

### Two main data structures:
1. **Series** = 1-dimensional (like a single column)
2. **DataFrame** = 2-dimensional (like a spreadsheet/table)

```python
# Let's start!
import pandas as pd
import numpy as np

print("Pandas version:", pd.__version__)
print("\nPandas is ready to use!")
print("pd = standard abbreviation for pandas")
```

## CELL 2: Creating a Series (1-Dimensional)

A **Series** is like a single column in Excel.

**Syntax:** `pd.Series(data, index=None)`

```python
# Method 1: From a list
grades = pd.Series([85, 90, 78, 92, 88])
print("Series from list:")
print(grades)
print("\nType:", type(grades))

# Notice: Pandas automatically creates an index (0, 1, 2, 3, 4)
```

```python
# Method 2: With custom index
grades_named = pd.Series(
    [85, 90, 78, 92, 88],
    index=['Math', 'Physics', 'Chemistry', 'English', 'CS']
)

print("Series with custom index:")
print(grades_named)

# Now you can access by name!
print("\nMath grade:", grades_named['Math'])
print("CS grade:", grades_named['CS'])
```

```python
# Method 3: From a dictionary (automatic index from keys)
student_scores = {
    'Tuvya': 85,
    'Tvisha': 92,
    'Rahul': 78,
    'Priya': 90
}

scores_series = pd.Series(student_scores)
print("Series from dictionary:")
print(scores_series)

print("\nTvisha's score:", scores_series['Tvisha'])
```

## CELL 3: Series Basic Operations

```python
# Using the grades_named Series from before
print("Original grades:")
print(grades_named)

# Mathematical operations (vectorized - like NumPy!)
print("\n+5 bonus points to everyone:")
print(grades_named + 5)

print("\nDouble all grades (just for fun):")
print(grades_named * 2)

# Statistical operations
print("\n=" * 50)
print("STATISTICS")
print("=" * 50)
print(f"Mean (average): {grades_named.mean():.2f}")
print(f"Median: {grades_named.median():.2f}")
print(f"Standard deviation: {grades_named.std():.2f}")
print(f"Minimum: {grades_named.min()}")
print(f"Maximum: {grades_named.max()}")
print(f"Sum: {grades_named.sum()}")
```

## CELL 4: Creating a DataFrame (2-Dimensional)

A **DataFrame** is like an Excel spreadsheet - rows and columns.

**This is what you'll use 90% of the time in real work.**

```python
# Method 1: From a dictionary of lists
students_data = {
    'Name': ['Tuvya', 'Tvisha', 'Rahul', 'Priya', 'Arjun'],
    'Age': [18, 18, 19, 18, 19],
    'Math': [85, 92, 78, 90, 88],
    'Physics': [80, 88, 82, 85, 90],
    'Chemistry': [88, 90, 75, 92, 85]
}

df = pd.DataFrame(students_data)
print("DataFrame created:")
print(df)

print("\nType:", type(df))
print("Shape (rows, columns):", df.shape)
```

## CELL 16: Handling Missing Data - Detection

```python
# Create DataFrame with missing values
data_missing = {
    'Name': ['Tuvya', 'Tvisha', 'Rahul', 'Priya', 'Arjun'],
    'Age': [18, 18, None, 18, 19],
    'Math': [85, None, 78, 90, 88],
    'Physics': [80, 88, 82, None, 90],
    'Email': ['tuvya@mail.com', None, 'rahul@mail.com', 'priya@mail.com', None]
}

df_missing = pd.DataFrame(data_missing)
print("DataFrame with missing values:")
print(df_missing)

# Detect missing values
print("\nisnull() - True where data is missing:")
print(df_missing.isnull())

print("\nnotnull() - True where data exists:")
print(df_missing.notnull())

# Count missing values per column
print("\nMissing values per column:")
print(df_missing.isnull().sum())

# Total missing values
print("\nTotal missing values:", df_missing.isnull().sum().sum())
```

## CELL 17: Handling Missing Data - Removal

```python
# dropna() - remove rows/columns with missing values

# Drop rows with ANY missing values
print("Drop rows with ANY missing value:")
df_dropped_rows = df_missing.dropna()
print(df_dropped_rows)

# Drop columns with ANY missing values
print("\nDrop columns with ANY missing value:")
df_dropped_cols = df_missing.dropna(axis=1)
print(df_dropped_cols)

# Drop rows where ALL values are missing
print("\nDrop rows only if ALL values are missing:")
df_dropped_all = df_missing.dropna(how='all')
print(df_dropped_all)

# Drop rows with missing values in specific columns
print("\nDrop rows with missing 'Math' or 'Physics':")
df_dropped_specific = df_missing.dropna(subset=['Math', 'Physics'])
print(df_dropped_specific)
```

## CELL 18: Handling Missing Data - Filling

```python
# fillna() - replace missing values

# Fill with a constant value
print("Fill all missing with 0:")
df_filled_zero = df_missing.fillna(0)
print(df_filled_zero)

# Fill with different values for different columns
print("\nFill missing values with column-specific values:")
df_filled_dict = df_missing.fillna({
    'Age': df_missing['Age'].mean(),  # Mean age
    'Math': df_missing['Math'].mean(),  # Mean math score
    'Physics': df_missing['Physics'].median(),  # Median physics
    'Email': 'unknown@mail.com'  # Default email
})
print(df_filled_dict)

# Forward fill (use previous value)
print("\nForward fill (ffill):")
df_ffill = df_missing.fillna(method='ffill')
print(df_ffill)

# Backward fill (use next value)
print("\nBackward fill (bfill):")
df_bfill = df_missing.fillna(method='bfill')
print(df_bfill)
```

## CELL 20: Working with Dates

```python
# Create DataFrame with date strings
dates_df = pd.DataFrame({
    'Date': ['2026-01-15', '2026-02-20', '2026-03-31', '2026-04-01'],
    'Event': ['Semester Start', 'Mid-terms', 'Met Tvisha', 'Transformation Start'],
    'Score': [0, 75, 80, 100]
})

print("Original (dates as strings):")
print(dates_df)
print("\nDate column type:", dates_df['Date'].dtype)

# Convert to datetime
dates_df['Date'] = pd.to_datetime(dates_df['Date'])
print("\nAfter conversion to datetime:")
print(dates_df)
print("Date column type:", dates_df['Date'].dtype)

# Extract date components
dates_df['Year'] = dates_df['Date'].dt.year
dates_df['Month'] = dates_df['Date'].dt.month
dates_df['Day'] = dates_df['Date'].dt.day
dates_df['DayOfWeek'] = dates_df['Date'].dt.day_name()

print("\nWith extracted components:")
print(dates_df)

# Date arithmetic
print("\n30 days after each date:")
dates_df['30_Days_Later'] = dates_df['Date'] + pd.Timedelta(days=30)
print(dates_df[['Event', 'Date', '30_Days_Later']])
```

## CELL 21: Apply Function

```python
# apply() - apply a function to each element/row/column

# Example: Custom grading function
def get_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'D'

# Apply to a column
df['Math_Grade'] = df['Math'].apply(get_grade)
print("Math grades:")
print(df[['Name', 'Math', 'Math_Grade']])

# Apply with lambda function (shorter)
df['Pass_Fail'] = df['Percentage'].apply(lambda x: 'Pass' if x >= 60 else 'Fail')
print("\nPass/Fail status:")
print(df[['Name', 'Percentage', 'Pass_Fail']])

# Apply to entire row
def calculate_average(row):
    return (row['Math'] + row['Physics'] + row['Chemistry']) / 3

df['Average'] = df.apply(calculate_average, axis=1)
print("\nWith average:")
print(df[['Name', 'Math', 'Physics', 'Chemistry', 'Average']])
```

## CELL 22: Map Function

```python
# map() - replace values based on mapping

# Create simple DataFrame
students = pd.DataFrame({
    'Name': ['Tuvya', 'Tvisha', 'Rahul'],
    'Grade': ['B', 'A', 'C']
})

print("Original:")
print(students)

# Map grades to GPA
grade_to_gpa = {
    'A': 4.0,
    'B': 3.0,
    'C': 2.0,
    'D': 1.0
}

students['GPA'] = students['Grade'].map(grade_to_gpa)
print("\nWith GPA:")
print(students)

# Map with function
students['Grade_Description'] = students['Grade'].map(
    lambda x: 'Excellent' if x == 'A' else 'Good' if x == 'B' else 'Average'
)
print("\nWith description:")
print(students)
```

## CELL 25: Conditional Column Creation

```python
# np.where() - vectorized if-else
fitness = pd.DataFrame({
    'Name': ['Tuvya', 'Tvisha', 'Rahul', 'Priya'],
    'Weight': [91, 55, 75, 60],
    'Height': [180, 165, 175, 160]  # in cm
})

print("Fitness data:")
print(fitness)

# Simple condition
fitness['Weight_Category'] = np.where(
    fitness['Weight'] > 70,
    'Heavy',
    'Light'
)
print("\nWith weight category:")
print(fitness)

# Multiple conditions with np.select()
conditions = [
    fitness['Weight'] > 80,
    fitness['Weight'] > 60,
    fitness['Weight'] <= 60
]
choices = ['Need to lose weight', 'Maintain', 'Gain weight']

fitness['Goal'] = np.select(conditions, choices, default='Unknown')
print("\nWith goals:")
print(fitness)
```

## CELL 26: Binning (Categorizing Continuous Data)

```python
# cut() - bin continuous data into discrete intervals

scores = pd.DataFrame({
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Score': [45, 67, 89, 92, 78, 55, 98, 71]
})

print("Scores:")
print(scores)

# Create bins
bins = [0, 60, 75, 90, 100]
labels = ['Fail', 'Pass', 'Good', 'Excellent']

scores['Grade'] = pd.cut(scores['Score'], bins=bins, labels=labels)
print("\nWith grades:")
print(scores)

# Equal-width bins (automatic)
scores['Quartile'] = pd.cut(scores['Score'], bins=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
print("\nWith quartiles:")
print(scores)

# qcut() - quantile-based binning (equal-sized bins)
scores['Percentile'] = pd.qcut(scores['Score'], q=4, labels=['Bottom 25%', '25-50%', '50-75%', 'Top 25%'])
print("\nWith percentiles:")
print(scores)
```

## CELL 31: GroupBy - Basic Aggregation

```python
# Create sales data
sales = pd.DataFrame({
    'Region': ['North', 'South', 'North', 'East', 'South', 'North', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B'],
    'Sales': [100, 150, 120, 200, 180, 90, 210, 140],
    'Profit': [20, 30, 25, 50, 40, 15, 55, 28]
})

print("Sales data:")
print(sales)

# Group by single column
print("\nTotal sales by region:")
regional_sales = sales.groupby('Region')['Sales'].sum()
print(regional_sales)

# Multiple aggregations
print("\nRegional statistics:")
regional_stats = sales.groupby('Region')['Sales'].agg(['sum', 'mean', 'count'])
print(regional_stats)

# Group by multiple columns
print("\nSales by region and product:")
region_product = sales.groupby(['Region', 'Product'])['Sales'].sum()
print(region_product)
```

## CELL 32: GroupBy - Multiple Aggregations

```python
# Different aggregations for different columns
print("Comprehensive regional analysis:")
analysis = sales.groupby('Region').agg({
    'Sales': ['sum', 'mean', 'max'],
    'Profit': ['sum', 'mean']
})
print(analysis)

# Custom aggregation function
def profit_margin(x):
    return (x['Profit'].sum() / x['Sales'].sum()) * 100

print("\nProfit margin by region:")
margins = sales.groupby('Region').apply(profit_margin)
print(margins)

# Reset index to make result a DataFrame
print("\nAs DataFrame:")
margins_df = margins.reset_index(name='Profit_Margin_%')
print(margins_df)
```

## CELL 33: Merging DataFrames - Inner Join

```python
# Create two related DataFrames
students_info = pd.DataFrame({
    'Student_ID': [1, 2, 3, 4],
    'Name': ['Tuvya', 'Tvisha', 'Rahul', 'Priya'],
    'Age': [18, 18, 19, 18]
})

students_scores = pd.DataFrame({
    'Student_ID': [1, 2, 3, 5],
    'Math': [85, 92, 78, 88],
    'Physics': [80, 88, 82, 90]
})

print("Students info:")
print(students_info)
print("\nStudents scores:")
print(students_scores)

# Inner join (only matching records)
print("\nInner join (only students in both tables):")
merged_inner = pd.merge(students_info, students_scores, on='Student_ID', how='inner')
print(merged_inner)
print(f"Result has {len(merged_inner)} rows (only IDs 1, 2, 3 match)")
```

## CELL 34: Merging DataFrames - Other Joins

```python
# Left join (all from left table)
print("Left join (all students from info table):")
merged_left = pd.merge(students_info, students_scores, on='Student_ID', how='left')
print(merged_left)
print("Notice: Priya (ID 4) has NaN scores because she's not in scores table")

# Right join (all from right table)
print("\nRight join (all students from scores table):")
merged_right = pd.merge(students_info, students_scores, on='Student_ID', how='right')
print(merged_right)
print("Notice: Student ID 5 has NaN name/age because not in info table")

# Outer join (all from both tables)
print("\nOuter join (all students from both tables):")
merged_outer = pd.merge(students_info, students_scores, on='Student_ID', how='outer')
print(merged_outer)
print("Notice: Both Priya and ID 5 are included with NaN where data is missing")
```

## CELL 36: Pivot Tables

## CELL 37: Reshaping - Melt (Wide to Long)

## CELL 39: Window Functions - Rolling

```python
# Time series data (e.g., daily scores)
daily_scores = pd.DataFrame({
    'Day': range(1, 11),
    'Score': [85, 88, 82, 90, 87, 92, 89, 91, 86, 93]
})

print("Daily scores:")
print(daily_scores)

# 3-day moving average
daily_scores['MA_3'] = daily_scores['Score'].rolling(window=3).mean()
print("\nWith 3-day moving average:")
print(daily_scores)

# 3-day moving sum
daily_scores['Sum_3'] = daily_scores['Score'].rolling(window=3).sum()
print("\nWith 3-day sum:")
print(daily_scores)

# 3-day max
daily_scores['Max_3'] = daily_scores['Score'].rolling(window=3).max()
print("\nWith 3-day maximum:")
print(daily_scores)
```

## CELL 40: Window Functions - Expanding & Cumulative

```python
# Expanding (cumulative from start)
daily_scores['Expanding_Mean'] = daily_scores['Score'].expanding().mean()
print("Expanding mean (cumulative average):")
print(daily_scores[['Day', 'Score', 'Expanding_Mean']])

# Cumulative sum
daily_scores['Cumsum'] = daily_scores['Score'].cumsum()
print("\nCumulative sum:")
print(daily_scores[['Day', 'Score', 'Cumsum']])

# Cumulative max
daily_scores['Cummax'] = daily_scores['Score'].cummax()
print("\nCumulative maximum:")
print(daily_scores[['Day', 'Score', 'Cummax']])
```

## CELL 41: Rank and Percentile

## CELL 44: Performance Optimization Tips

```python
# Create large DataFrame for demonstration
large_df = pd.DataFrame({
    'A': np.random.randint(0, 100, 100000),
    'B': np.random.randn(100000),
    'C': np.random.choice(['X', 'Y', 'Z'], 100000)
})

print(f"Large DataFrame: {large_df.shape}")
print(f"Memory usage: {large_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# TIP 1: Use categorical for repeated string values
large_df['C_cat'] = large_df['C'].astype('category')
print(f"\nAfter converting to categorical:")
print(f"Memory usage: {large_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# TIP 2: Use appropriate dtypes
# int64 can be downcasted to int32 or int8 if values are small
print(f"\nColumn A original dtype: {large_df['A'].dtype}")
large_df['A'] = large_df['A'].astype('int8')  # Values are 0-100, int8 is enough
print(f"Column A new dtype: {large_df['A'].dtype}")
print(f"Memory usage: {large_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# TIP 3: Use vectorized operations instead of loops
# SLOW (don't do this):
# for i in range(len(df)):
#     df.loc[i, 'new_col'] = df.loc[i, 'A'] * 2

# FAST (do this):
large_df['new_col'] = large_df['A'] * 2

print("\nOptimization tips:")
print("1. Use categorical for repeated strings")
print("2. Downcast numeric types when possible")
print("3. Use vectorized operations, avoid loops")
print("4. Use chunksize for very large files")
print("5. Use .loc and .iloc instead of .ix (deprecated)")
```

## CELL 45: Chaining Operations (Method Chaining)

```python
# Create messy data
messy = pd.DataFrame({
    'name': ['TUVYA', 'tvisha', 'RaHuL', 'PRIYA', 'tvisha'],
    'age': [18, 18, 19, 18, 18],
    'score': [85, 92, 78, 90, 92]
})

print("Messy data:")
print(messy)

# Method chaining - clean in one go
clean = (
    messy
    .drop_duplicates()                                    # Remove duplicates
    .assign(name=lambda x: x['name'].str.title())        # Title case names
    .assign(grade=lambda x: pd.cut(x['score'],           # Add grade
                                    bins=[0, 75, 90, 100],
                                    labels=['C', 'B', 'A']))
    .sort_values('score', ascending=False)                # Sort by score
    .reset_index(drop=True)                               # Reset index
)

print("\nCleaned data (one chain):")
print(clean)

# Same thing without chaining (harder to read)
# clean = messy.drop_duplicates()
# clean['name'] = clean['name'].str.title()
# clean['grade'] = pd.cut(clean['score'], bins=[0, 75, 90, 100], labels=['C', 'B', 'A'])
# clean = clean.sort_values('score', ascending=False)
# clean = clean.reset_index(drop=True)

print("\nMethod chaining makes code cleaner and more readable!")
```

## CELL 46: APPLICATION 1 - Student Performance Analysis

**Scenario:** You're a TA analyzing student performance across multiple subjects.

**Questions to answer:**
1. Which students are at risk (average < 70)?
2. Which subject has lowest average?
3. Who are the top 3 performers?
4. What's the grade distribution?

```python
# Create student dataset
np.random.seed(42)
students = pd.DataFrame({
    'Student_ID': range(1, 21),
    'Name': [f'Student_{i}' for i in range(1, 21)],
    'Math': np.random.randint(50, 100, 20),
    'Physics': np.random.randint(50, 100, 20),
    'Chemistry': np.random.randint(50, 100, 20),
    'Biology': np.random.randint(50, 100, 20),
    'English': np.random.randint(50, 100, 20)
})

print("=" * 80)
print("STUDENT PERFORMANCE ANALYSIS")
print("=" * 80)

print("\nDataset (first 10 students):")
print(students.head(10))
print(f"\nTotal students: {len(students)}")
```

```python
# Calculate average per student
subject_cols = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']
students['Average'] = students[subject_cols].mean(axis=1)
students['Total'] = students[subject_cols].sum(axis=1)

# Q1: Students at risk (average < 70)
print("=" * 80)
print("QUESTION 1: Students at risk (Average < 70)")
print("=" * 80)
at_risk = students[students['Average'] < 70][['Name', 'Average', 'Total']].sort_values('Average')
print(at_risk)
print(f"\n{len(at_risk)} students need attention")
```

```python
# Q2: Subject with lowest average
print("=" * 80)
print("QUESTION 2: Subject Performance")
print("=" * 80)
subject_avg = students[subject_cols].mean().sort_values()
print("\nAverage scores by subject:")
print(subject_avg)

print(f"\nLowest performing subject: {subject_avg.idxmin()} ({subject_avg.min():.2f})")
print(f"Highest performing subject: {subject_avg.idxmax()} ({subject_avg.max():.2f})")
```

```python
# Q3: Top 3 performers
print("=" * 80)
print("QUESTION 3: Top 3 Performers")
print("=" * 80)
top_3 = students.nlargest(3, 'Average')[['Name', 'Average', 'Total'] + subject_cols]
print(top_3)

# Add rank
students['Rank'] = students['Average'].rank(ascending=False, method='min')
print("\nTop 10 Leaderboard:")
print(students.nsmallest(10, 'Rank')[['Rank', 'Name', 'Average', 'Total']])
```

```python
# Q4: Grade distribution
print("=" * 80)
print("QUESTION 4: Grade Distribution")
print("=" * 80)

# Assign grades
students['Grade'] = pd.cut(
    students['Average'],
    bins=[0, 60, 70, 80, 90, 100],
    labels=['F', 'D', 'C', 'B', 'A']
)

grade_dist = students['Grade'].value_counts().sort_index()
print("\nGrade distribution:")
print(grade_dist)

print("\nGrade distribution (percentage):")
print((grade_dist / len(students) * 100).round(1))

# Summary statistics
print("\n" + "=" * 80)
print("OVERALL STATISTICS")
print("=" * 80)
print(f"Class average: {students['Average'].mean():.2f}")
print(f"Class median: {students['Average'].median():.2f}")
print(f"Standard deviation: {students['Average'].std():.2f}")
print(f"Highest score: {students['Average'].max():.2f} ({students.loc[students['Average'].idxmax(), 'Name']})")
print(f"Lowest score: {students['Average'].min():.2f} ({students.loc[students['Average'].idxmin(), 'Name']})")
```

## CELL 47: APPLICATION 2 - E-commerce Sales Dashboard

**Scenario:** You're analyzing an e-commerce company's sales data.

**Questions:**
1. Total revenue by category and region
2. Best-selling products
3. Monthly sales trends
4. Customer purchase analysis

```python
# Create e-commerce dataset
np.random.seed(42)
n_orders = 500

ecommerce = pd.DataFrame({
    'Order_ID': range(1, n_orders + 1),
    'Date': pd.date_range('2026-01-01', periods=n_orders, freq='12H'),
    'Customer_ID': np.random.randint(1, 100, n_orders),
    'Product': np.random.choice(['Laptop', 'Phone', 'Tablet', 'Watch', 'Headphones'], n_orders),
    'Category': np.random.choice(['Electronics', 'Accessories'], n_orders, p=[0.7, 0.3]),
    'Region': np.random.choice(['North', 'South', 'East', 'West'], n_orders),
    'Quantity': np.random.randint(1, 5, n_orders),
    'Price': np.random.randint(100, 2000, n_orders)
})

# Calculate revenue
ecommerce['Revenue'] = ecommerce['Quantity'] * ecommerce['Price']

print("=" * 80)
print("E-COMMERCE SALES DASHBOARD")
print("=" * 80)
print(f"\nAnalyzing {len(ecommerce)} orders")
print(f"Date range: {ecommerce['Date'].min().date()} to {ecommerce['Date'].max().date()}")
print("\nSample data:")
print(ecommerce.head(10))
```

```python
# Q1: Revenue by category and region
print("=" * 80)
print("QUESTION 1: Revenue by Category and Region")
print("=" * 80)

revenue_analysis = ecommerce.pivot_table(
    values='Revenue',
    index='Category',
    columns='Region',
    aggfunc='sum',
    margins=True
)
print(revenue_analysis)

# Best performing region overall
regional_revenue = ecommerce.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
print(f"\nBest region: {regional_revenue.idxmax()} (₹{regional_revenue.max():,.0f})")

# Best performing category
category_revenue = ecommerce.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
print(f"Best category: {category_revenue.idxmax()} (₹{category_revenue.max():,.0f})")
```

```python
# Q2: Best-selling products
print("=" * 80)
print("QUESTION 2: Product Performance")
print("=" * 80)

product_stats = ecommerce.groupby('Product').agg({
    'Revenue': 'sum',
    'Quantity': 'sum',
    'Order_ID': 'count'
}).rename(columns={'Order_ID': 'Orders'})

product_stats['Avg_Order_Value'] = product_stats['Revenue'] / product_stats['Orders']
product_stats = product_stats.sort_values('Revenue', ascending=False)

print("\nProduct performance:")
print(product_stats)

print(f"\nBest seller (revenue): {product_stats.index[0]}")
print(f"Best seller (quantity): {product_stats.sort_values('Quantity', ascending=False).index[0]}")
print(f"Highest AOV: {product_stats.sort_values('Avg_Order_Value', ascending=False).index[0]}")
```

```python
# Q3: Monthly sales trends
print("=" * 80)
print("QUESTION 3: Monthly Trends")
print("=" * 80)

# Extract month
ecommerce['Month'] = ecommerce['Date'].dt.to_period('M')

monthly_sales = ecommerce.groupby('Month').agg({
    'Revenue': 'sum',
    'Order_ID': 'count',
    'Customer_ID': 'nunique'
}).rename(columns={'Order_ID': 'Orders', 'Customer_ID': 'Unique_Customers'})

monthly_sales['Avg_Order_Value'] = monthly_sales['Revenue'] / monthly_sales['Orders']

print("\nMonthly performance:")
print(monthly_sales)

# Growth rate
monthly_sales['Revenue_Growth_%'] = monthly_sales['Revenue'].pct_change() * 100
print("\nWith growth rate:")
print(monthly_sales)
```

```python
# Q4: Customer analysis
print("=" * 80)
print("QUESTION 4: Customer Analysis")
print("=" * 80)

customer_analysis = ecommerce.groupby('Customer_ID').agg({
    'Revenue': 'sum',
    'Order_ID': 'count',
    'Quantity': 'sum'
}).rename(columns={'Order_ID': 'Orders'})

customer_analysis['Avg_Order_Value'] = customer_analysis['Revenue'] / customer_analysis['Orders']

# Top 10 customers by revenue
print("\nTop 10 customers by revenue:")
top_customers = customer_analysis.nlargest(10, 'Revenue')
print(top_customers)

# Customer segmentation
customer_analysis['Segment'] = pd.cut(
    customer_analysis['Revenue'],
    bins=[0, 5000, 15000, 100000],
    labels=['Bronze', 'Silver', 'Gold']
)

print("\nCustomer segments:")
print(customer_analysis['Segment'].value_counts())

# Summary
print("\n" + "=" * 80)
print("EXECUTIVE SUMMARY")
print("=" * 80)
print(f"Total Revenue: ₹{ecommerce['Revenue'].sum():,.0f}")
print(f"Total Orders: {len(ecommerce):,}")
print(f"Unique Customers: {ecommerce['Customer_ID'].nunique()}")
print(f"Average Order Value: ₹{ecommerce['Revenue'].mean():,.2f}")
print(f"Best Product: {product_stats.index[0]}")
print(f"Best Region: {regional_revenue.idxmax()}")
```
